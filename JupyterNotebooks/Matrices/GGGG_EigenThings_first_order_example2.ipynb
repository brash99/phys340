# Coupled First Order ODEs

Consider the following system of coupled first order ODEs:

$$\begin{equation}
\frac{dx}{dt} =  4x + 2y\\
\frac{dy}{dt} =  -x + y
\end{equation}$$

where $x$ and $y$ are functions of time $t$.  This system of ODEs can be written in matrix form as:

$$\begin{equation}
\begin{bmatrix}
\frac{dx}{dt}\\
\frac{dy}{dt}
\end{bmatrix}
=
\begin{bmatrix}
4 & 2\\
-1 & 1
\end{bmatrix}
\begin{bmatrix}
x\\
y
\end{bmatrix}
\end{equation}$$

From the fundamental theorem for eigenvalues and eigenvectors, we know that we can write:

$$\begin{bmatrix} \frac{dx}{dt}\\ \frac{dy}{dt} \end{bmatrix} = S \Lambda S^{-1} \begin{bmatrix} x\\ y \end{bmatrix}$$


$$S^{-1} \begin{bmatrix} \frac{dx}{dt}\\ \frac{dy}{dt} \end{bmatrix} = \Lambda S^{-1} \begin{bmatrix} x\\ y \end{bmatrix}$$

We can define a new set of variables, $r_1$ and $r_2$, as:

$$\begin{equation}

\begin{bmatrix} r_1\\ r_2 \end{bmatrix} = S^{-1} \begin{bmatrix} x\\ y \end{bmatrix}

\end{equation}$$

Then, the system of ODEs can be written as

$$\begin{equation}

\begin{bmatrix} \frac{dr_1}{dt}\\ \frac{dr_2}{dt} \end{bmatrix} = \Lambda \begin{bmatrix} r_1\\ r_2 \end{bmatrix}

\end{equation}$$

where $\Lambda$ is a diagonal matrix with the eigenvalues of the coefficient matrix on the diagonal.

 

```python
import numpy as np

# Initial conditions

y10 = 1
y20 = 0

# create a coefficient matrix
A = np.array([[4, 2],[-1, 1]])
print(A)

# print the eigenvalues and eigenvectors of the coefficient matrix
eigenvalues, eigenvectorsT = np.linalg.eig(A)

# the eigenvectors are the columns of the eigenvectorsT matrix
# taking the transpose of this matrix gives the eigenvectors as rows
eigenvectors = eigenvectorsT.T
print('The eigenvalues are ')
print(eigenvalues)
print('The eigenvectors (columns) are ')
print(eigenvectorsT)
print()
```

## Solution of the System of ODEs

Solving for the eigenvalues and eigenvectors of the coefficient matrix, we find that the eigenvalues are $\lambda_1 = 3$ and $\lambda_2 = 2$. 

We can show this by solving the characteristic equation of the coefficient matrix:

$$\begin{equation}
\begin{vmatrix}
4-\lambda & 2\\
-1 & 1-\lambda
\end{vmatrix}
= (4-\lambda)(1-\lambda) + 2  = 0
\end{equation}$$

This leads to the characteristic equation $\lambda^2 - 5\lambda + 6 = 0$, which has solutions $\lambda_1 = 3$ and $\lambda_2 = 2$. 

The corresponding eigenvectors are $\mathbf{v}_1 = \begin{bmatrix}\frac{2}{\sqrt{5}}\\-\frac{1}{\sqrt{5}}\end{bmatrix}$ and $\mathbf{v}_2 = \begin{bmatrix}-\frac{1}{\sqrt{2}}\\ \frac{1}{\sqrt{2}}\end{bmatrix}$.

We can show this by solving the equation $(A - \lambda I)\mathbf{v} = 0$ for each eigenvalue.  For $\lambda_1 = 3$, we have

$$\begin{equation}
\begin{bmatrix}
4-3 & 2\\
-1 & 1-3
\end{bmatrix}
\begin{bmatrix}
v_{11}\\
v_{21}
\end{bmatrix}
= \begin{bmatrix}
1 & 2\\
-1 & -2
\end{bmatrix}
\begin{bmatrix}
v_{11}\\
v_{21}
\end{bmatrix}
= \begin{bmatrix}
0\\
0
\end{bmatrix}
\end{equation}$$

This leads to the equations $v_{11} + 2 v_{21} = 0$ and $-v_{11} - 2v_{21} = 0$.  Each of these leads to the condition $v_{21} = -\frac{1}{2}v_{11}$.

We can also impose that the eigenvector is normalized, so that $v_{11}^2 + v_{21}^2 = 1$.  This leads to the condition $v_{11}^2 + \frac{1}{4}v_{11}^2 = 1$, or $5v_{11}^2 = 4$, or $v_{11} = \frac{2}{\sqrt{5}}$.  Then, $v_{21} = -\frac{1}{2}v_{11} = -\frac{1}{\sqrt{5}}$.

For $\lambda_2 = 2$, we have

$$\begin{equation}
\begin{bmatrix}
4-2 & 2\\
-1 & 1-2
\end{bmatrix}
\begin{bmatrix}
v_{12}\\
v_{22}
\end{bmatrix}
= \begin{bmatrix}
2 & 2\\
-1 & -1
\end{bmatrix}
\begin{bmatrix}
v_{12}\\
v_{22}
\end{bmatrix}
= \begin{bmatrix}
0\\
0
\end{bmatrix}
\end{equation}$$

This leads to the equations $2v_{12} + 2v_{22} = 0$ and $-v_{12} - v_{22} = 0$.  Each of these leads to the condition $v_{22} = -v_{12}$.

We can also impose that the eigenvector is normalized, so that $v_{12}^2 + v_{22}^2 = 1$.  This leads to the condition $v_{12}^2 + v_{12}^2 = 1$, or $2v_{12}^2 = 1$, or $v_{12} = \frac{1}{\sqrt{2}}$.  Then, $v_{22} = -v_{12} = -\frac{1}{\sqrt{2}}$.

The eigenvectors are then $\mathbf{v}_1 = \begin{bmatrix}\frac{2}{\sqrt{5}}\\-\frac{1}{\sqrt{5}}\end{bmatrix}$ and $\mathbf{v}_2 = \begin{bmatrix}\frac{1}{\sqrt{2}}\\-\frac{1}{\sqrt{2}}\end{bmatrix}$.

From these, we create a similar matrix $S$ with the eigenvectors as columns, and a diagonal matrix $L$ with the eigenvalues on the diagonal:

$$\begin{equation}
S = \begin{bmatrix}
\frac{2}{\sqrt{5}} & -\frac{1}{\sqrt{2}}\\
-\frac{1}{\sqrt{5}} & \frac{1}{\sqrt{2}}
\end{bmatrix}
\end{equation}$$

$$\begin{equation}
L = \begin{bmatrix}
3& 0\\
0 & 2
\end{bmatrix}
\end{equation}$$

The determinant of the matrix $S$ is $\frac{2}{\sqrt{5}}(\frac{1}{\sqrt{2}}) - \frac{-1}{\sqrt{5}}(\frac{-1}{\sqrt{2}}) = \frac{1}{\sqrt{10}}$.  

The inverse of the matrix $S$ is then

$$\begin{equation}
S^{-1} = \frac{1}{\frac{1}{\sqrt{10}}}\begin{bmatrix}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}\\
\frac{1}{\sqrt{5}} & \frac{2}{\sqrt{5}}
\end{bmatrix}
\end{equation}$$

or, simplifying, 

$$\begin{equation}
S^{-1} = \begin{bmatrix}
\sqrt{5} & \sqrt{5} \\
\sqrt{2} & 2 \sqrt{2}
\end{bmatrix}
\end{equation}$$

The solution of the system of ODEs is then

$$\begin{equation}
\begin{bmatrix}
x\\
y
\end{bmatrix}
=
S
\begin{bmatrix}
r_1\\
r_2
\end{bmatrix}
=
\begin{bmatrix}
\frac{2}{\sqrt{5}} & -\frac{1}{\sqrt{2}}\\
-\frac{1}{\sqrt{5}} & \frac{1}{\sqrt{2}}
\end{bmatrix}
\begin{bmatrix}
r_1\\
r_2
\end{bmatrix}
\end{equation}$$

where $r_1$ and $r_2$ are the solutions of the system of ODEs

$$\begin{equation}
\begin{bmatrix}
\frac{dr_1}{dt}\\
\frac{dr_2}{dt}
\end{bmatrix}
=
\begin{bmatrix}
3& 0\\
0 & 2
\end{bmatrix}
\begin{bmatrix}
r_1\\
r_2
\end{bmatrix}
\end{equation}$$

The solution of this system of ODEs is
    
$$\begin{equation}
\begin{bmatrix}
r_1\\
r_2
\end{bmatrix}
=
\begin{bmatrix}
r_{10} e^{3t}\\
r_{20} e^{2t}
\end{bmatrix}
\end{equation}$$

where $r_{10}$, $r_{20}$  are constants of integration.

We can determine these constants of integration from the initial conditions.  If the initial conditions are $x(0) = 1$, $y(0) =0$ then the solution of the system of ODEs is

$$\begin{equation}
\begin{bmatrix}
r_{10}\\
r_{20}
\end{bmatrix}
=
\begin{bmatrix}
\sqrt{5} & \sqrt{5}\\
\sqrt{2} & 2\sqrt{2}
\end{bmatrix}
\begin{bmatrix}
1\\
0
\end{bmatrix}
\end{equation}$$


This leads to the following:

$$\begin{equation}
r_{10}= \sqrt{5}  \\
r_{20}= \sqrt{2}
\end{equation}$$

The solution of the system of ODEs is then

$$\begin{equation}
\begin{bmatrix}
x\\
y
\end{bmatrix}
=
S
\begin{bmatrix}
r\\
s
\end{bmatrix}
=
\begin{bmatrix}
\frac{2}{\sqrt{5}} & -\frac{1}{\sqrt{2}}\\
-\frac{1}{\sqrt{5}} & \frac{1}{\sqrt{2}}
\end{bmatrix}
\begin{bmatrix}
\sqrt{5} e^{3t}\\
\sqrt{2} e^{2t}
\end{bmatrix}
\end{equation}$$

Multiplying the matrices, we find

$$\begin{equation}
\begin{bmatrix}
x\\
y
\end{bmatrix}
=
\begin{bmatrix}
2e^{3t} - e^{2t})\\
- e^{3t} + e^{2t})
\end{bmatrix}
\end{equation}$$


```python
import sympy as sym

# create a solution matrix
R = sym.Matrix([sym.symbols('r1'), sym.symbols('r2')])
sym.pretty_print(R)

# create a matrix of eigenvectors
S = sym.Matrix(eigenvectorsT)
sym.pretty_print(S)

# create a diagonal matrix of eigenvalues
L = sym.Matrix([[eigenvalues[0], 0], [0, eigenvalues[1]]])

# create the inverse of the matrix of eigenvectors
Sinv = S.inv()
sym.pretty_print(Sinv)

r1 = sym.symbols('r10')*sym.exp(eigenvalues[0] * sym.symbols('t'))
r2 = sym.symbols('r20')*sym.exp(eigenvalues[1] * sym.symbols('t'))

sym.pretty_print(r1)
sym.pretty_print(r2)

Y = S * R

print('The solution is')
sym.pretty_print(Y)

print('r1(0) = ',r1.subs(sym.symbols('t'),0).subs(sym.symbols('t'),0))
print('r2(0) = ',r2.subs(sym.symbols('t'),0).subs(sym.symbols('t'),0))


```

```python
Y0 = S * R.subs(sym.symbols('r1'),r1).subs(sym.symbols('r2'),r2)
Y0 = Y0.subs(sym.symbols('t'),0)
sym.pretty_print(Y0)

Y0dot = S * R.subs(sym.symbols('r1'),sym.diff(r1,sym.symbols('t'))).subs(sym.symbols('r2'),sym.diff(r2,sym.symbols('t')))
Y0dot = Y0dot.subs(sym.symbols('t'),0)
sym.pretty_print(Y0dot)

Y0val = sym.Matrix([sym.symbols('y10'), sym.symbols('y20')])
sym.pretty_print(Y0val)

Y0val = Y0val.subs(sym.symbols('y10'),y10).subs(sym.symbols('y20'),y20)

sym.pretty_print(Y0val)
```

```python
constant_dict = sym.solve([Y0val[0] - Y0[0], Y0val[1] - Y0[1]], [sym.symbols('r10'), sym.symbols('r20')])
```

```python
print(constant_dict[sym.symbols('r10')])
print(constant_dict[sym.symbols('r20')])
```

```python
Y_solution = Y.subs(sym.symbols('r1'),r1).subs(sym.symbols('r2'),r2)
sym.pretty_print(Y_solution)
Y_solution = Y_solution.subs(constant_dict)
sym.pretty_print(Y_solution)
```

```python
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 100)

xt = sym.lambdify(sym.symbols('t'), Y_solution[0], 'numpy')
yt = sym.lambdify(sym.symbols('t'), Y_solution[1], 'numpy')

plt.plot(t, xt(t), label='x')
plt.plot(t, yt(t), label='y')
plt.legend()
plt.show()
```

```python
# now, let's calculate the motion numerically, using the Euler method
# time step
dt = 0.001

# number of time steps
n = 1000

# create arrays to store the solution
tn = np.zeros(n)
x = np.zeros(n)
y = np.zeros(n)
xdot = np.zeros(n)
ydot = np.zeros(n)

# initial conditions
tn[0] = 0
x[0] = 1
y[0] = 0

# loop over time steps
for i in range(n-1):
    # calculate the derivatives
    xdot[i] = 4*x[i] + 2*y[i]
    ydot[i] = -1*x[i] + y[i]
    
    # calculate the new values of y1 and y2 (Euler method)
    x[i+1] = x[i] + xdot[i] * dt
    y[i+1] = y[i] + ydot[i] * dt
    
    # calculate the new time
    tn[i+1] = tn[i] + dt
```

```python
# plot the results
fig, ax = plt.subplots(2,figsize=(10, 10))
ax[0].plot(tn, x, 'r.', label='x')
ax[1].plot(tn, y, 'r.', label='y')
ax[0].plot(t, xt(t), 'b-', label='x analytical', linewidth=4)
ax[1].plot(t, yt(t), 'b-', label='y analytical', linewidth=4)
ax[0].set_xlabel('t (s)')
ax[1].set_xlabel('t (s)')
ax[0].set_ylabel('x (m)')
ax[1].set_ylabel('y (m)')
ax[0].legend()
ax[1].legend()
plt.show()
```

Coupled First-order Differential Equations

Problem Setup

Consider a system of  n  coupled first-order linear differential equations:

$$\frac{d\mathbf{X}}{dt} = A\mathbf{X},$$

where:
	•	$\mathbf{X}$ is an  n -dimensional vector of dependent variables: $\mathbf{X} = [x_1(t), x_2(t), \ldots, x_n(t)]^T$,
	•	A is an  n \times n  matrix of constants, and
	•	$\frac{d\mathbf{X}}{dt}$ is the derivative of $\mathbf{X}$ with respect to t.

The goal is to solve for $\mathbf{X}(t)$.

Step 1: Find Eigenvalues and Eigenvectors of A

Compute the eigenvalues:
	
Solve the characteristic equation:

$\det(A - \lambda I) = 0,$

where $\lambda$ represents the eigenvalues and $I$ is the identity matrix.


Find the eigenvectors:

For each eigenvalue $\lambda_i$, solve:

$(A - \lambda_i I)\mathbf{v}_i = 0,$

to find the corresponding eigenvector $\mathbf{v}_i.$


Form the General Solution:

Diagonalize A (if possible):
	
If A is diagonalizable, construct a matrix $P$ from the eigenvectors:

$P = [\mathbf{v}_1 \, \mathbf{v}_2 \, \ldots \, \mathbf{v}_n],$

and a diagonal matrix \Lambda with eigenvalues on the diagonal:

$\Lambda = \text{diag}(\lambda_1, \lambda_2, \ldots, \lambda_n)$.

Then, A can be expressed as:

$A = P\Lambda P^{-1}.$


Solve the decoupled system:
	
Let $\mathbf{Y} = P^{-1}\mathbf{X}$. Substituting $\mathbf{X} = P\mathbf{Y}$ into the original equation gives:

$\frac{d\mathbf{Y}}{dt} = \Lambda\mathbf{Y}$.

Since $\Lambda$ is diagonal, this system is decoupled, and the equations can be solved independently:

$\frac{dy_i}{dt} = \lambda_i y_i \implies y_i(t) = C_i e^{\lambda_i t}$,

where $C_i$ are constants determined by initial conditions.


Transform back to X:
	
Using $\mathbf{X} = P\mathbf{Y}$, write:

$\mathbf{X}(t) = P \mathbf{Y}(t) = P \begin{bmatrix}
C_1 e^{\lambda_1 t} \\
C_2 e^{\lambda_2 t} \\
\vdots \\
C_n e^{\lambda_n t}
\end{bmatrix}$.


Step 3: Handle Non-Diagonalizable Cases (Jordan Form)

If A is not diagonalizable:

Compute the Jordan normal form $J$ such that $A = PJ P^{-1}$, where $J$ is a block matrix with eigenvalues on the diagonal and possibly non-zero entries directly above the diagonal.
	
Solve the modified system:

$\frac{d\mathbf{Y}}{dt} = J\mathbf{Y}$.

For each Jordan block, the solution will involve terms like $t e^{\lambda t}, t^2 e^{\lambda t}$, etc.

Step 4: Apply Initial Conditions

Use the given initial conditions $\mathbf{X}(0) = \mathbf{X}_0$ to find the constants $C_i$. Substitute $t = 0$ into the general solution:

$\mathbf{X}_0 = P \mathbf{C}$,

where $\mathbf{C} = [C_1, C_2, \ldots, C_n]^T$. Solve for $\mathbf{C}$:

$\mathbf{C} = P^{-1}\mathbf{X}_0$.

Summary of Final Solution

The final solution is:

$\mathbf{X}(t) = P \begin{bmatrix}
C_1 e^{\lambda_1 t} \\
C_2 e^{\lambda_2 t} \\
\vdots \\
C_n e^{\lambda_n t}
\end{bmatrix}$,

with C_i determined from initial conditions.

Example

Solve:

$\frac{d\mathbf{X}}{dt} = \begin{bmatrix}
2 & 1 \\
1 & 2
\end{bmatrix}\mathbf{X}, \quad \mathbf{X}(0) = \begin{bmatrix}
1 \\
0
\end{bmatrix}$.

Eigenvalues: $\lambda = 3, 1.$

Eigenvectors: For $\lambda = 3, \mathbf{v}_1 = \begin{bmatrix} 1 \\ 1 \end{bmatrix}; for \lambda = 1, \mathbf{v}_2 = \begin{bmatrix} 1 \\ -1 \end{bmatrix}.$

General solution:

$\mathbf{X}(t) = C_1 \begin{bmatrix} 1 \\ 1 \end{bmatrix} e^{3t} + C_2 \begin{bmatrix} 1 \\ -1 \end{bmatrix} e^{t}$.

Apply initial conditions: $\mathbf{X}(0) = \begin{bmatrix} 1 \\ 0 \end{bmatrix}$, solve for $C_1, C_2$.

Final solution:

$\mathbf{X}(t) = \frac{1}{2}\begin{bmatrix} 1 \\ 1 \end{bmatrix}e^{3t} + \frac{1}{2}\begin{bmatrix} 1 \\ -1 \end{bmatrix}e^{t}$.


This concludes the method!
