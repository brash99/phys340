# Stokes Theorem

In physics, Stokes' theorem is a statement about the integration of differential forms on manifolds, which both simplifies and generalizes several theorems from vector calculus. 

Stokes' theorem says that the integral of a differential form ω over the boundary of some orientable manifold Ω is equal to the integral of its exterior derivative dω over the whole of Ω. 

In its simplest form, Stokes' theorem says that the integral of a differential form ω over the boundary of some orientable manifold Ω is equal to the integral of its exterior derivative dω over the whole of Ω. 

In physics, Stokes' theorem is used in classical electrodynamics as the mathematical basis of electromagnetic potentials and in fluid mechanics as the mathematical basis of the Navier–Stokes equations. 

It is also used in Stokes flow, a mathematical model in fluid mechanics which is used to study the motion of very viscous fluids. Stokes' theorem is also used in the proof of the Atiyah–Singer index theorem, an important result in differential geometry and the theory of elliptic operators. 

Stokes' theorem is a special case of the more general Cartan–Stokes theorem, which in turn is a special case of the generalized Stokes' theorem. The theorem is named after the Irish mathematician and physicist George Gabriel Stokes (1819–1903), who published it at the University of Cambridge in 1850.

# Theorem

$$\int_{\partial \Omega} \omega = \int_{\Omega} d\omega$$

where $\omega$ is a differential form, $\Omega$ is a manifold with boundary $\partial \Omega$ and $d\omega$ is the exterior derivative of $\omega$.

## Example

Consider a vector field $\mathbf{F}$ defined as

$$\mathbf{F} = (x^2 + y^2) \mathbf{i} + (y^2 + z^2) \mathbf{j} + (z^2 + x^2) \mathbf{k}$$

and a surface $\Sigma$ defined as

$$\Sigma = \{ (x, y, z) \in \mathbb{R}^3 : x^2 + y^2 + z^2 = 1, z \geq 0 \}$$

The boundary of $\Sigma$ is the circle $x^2 + y^2 = 1, z = 0$.

The curl of $\mathbf{F}$ is given by the determinant of the following matrix:

$$\begin{vmatrix} \mathbf{i} & \mathbf{j} & \mathbf{k} \\ \frac{\partial}{\partial x} & \frac{\partial}{\partial y} & \frac{\partial}{\partial z} \\ x^2 + y^2 & y^2 + z^2 & z^2 + x^2 \end{vmatrix}$$

Evaluated at $(x, y, z)$, this is equal to

$$\left( \frac{\partial}{\partial y} (z^2 + x^2) - \frac{\partial}{\partial z} (y^2 + z^2) \right) \mathbf{i} - \left( \frac{\partial}{\partial x} (z^2 + x^2) - \frac{\partial}{\partial z} (x^2 + y^2) \right) \mathbf{j} + \left( \frac{\partial}{\partial x} (y^2 + z^2) - \frac{\partial}{\partial y} (x^2 + y^2) \right) \mathbf{k}$$

$$= -2(z \mathbf{i} + x \mathbf{j} + y \mathbf{k})$$

The integral of the curl of $\mathbf{F}$ over $\Sigma$ is

$$\int_{\Sigma} \nabla \times \mathbf{F} \cdot d\mathbf{S} = \int_{\Sigma} -2(z \mathbf{i} + x \mathbf{j} + y \mathbf{k}) \cdot d\mathbf{S}$$

where $d\mathbf{S}$ is the outward-pointing unit normal vector to $\Sigma$.  This normal vector is given by

$$d\mathbf{S} = \frac{\nabla (x^2 + y^2 + z^2)}{\| \nabla (x^2 + y^2 + z^2) \|} = \frac{2x \mathbf{i} + 2y \mathbf{j} + 2z \mathbf{k}}{2\sqrt{x^2 + y^2 + z^2}} = \frac{x \mathbf{i} + y \mathbf{j} + z \mathbf{k}}{\sqrt{x^2 + y^2 + z^2}}$$

Thus, the integral of the curl of $\mathbf{F}$ over $\Sigma$ is

$$\int_{\Sigma} \nabla \times \mathbf{F} \cdot d\mathbf{S} = \int_{\Sigma} -2(z \mathbf{i} + x \mathbf{j} + y \mathbf{k}) \cdot \frac{x \mathbf{i} + y \mathbf{j} + z \mathbf{k}}{\sqrt{x^2 + y^2 + z^2}}$$

$$= \int_{\Sigma} -2(xz + yz + xy) \frac{1}{\sqrt{x^2 + y^2 + z^2}}$$

$$= \int_{\Sigma} -2(xz + yz + xy) $$

since the radius of $\Sigma$ is 1.  Further, we can parameterize $\Sigma$ as

$$\mathbf{r}(t, \theta) = \cos t \sin \theta \mathbf{i} + \sin t \sin \theta \mathbf{j} + \cos \theta \mathbf{k}$$

where $0 \leq t \leq 2\pi$ and $0 \leq \theta \leq \pi/2$.  Then

$$\int_{\Sigma} \nabla \times \mathbf{F} \cdot d\mathbf{S} = \int_0^{2\pi} \int_0^{\pi/2} -2(\cos t \sin \theta \cos \theta + \sin t \sin \theta \cos \theta + \cos t \sin t \sin^2 \theta) \sin \theta d\theta dt$$

$$= \int_0^{2\pi} \int_0^{\pi/2} -2(\cos t \sin^2 \theta \cos \theta + \sin t \sin^2 \theta \cos \theta + \cos t \sin t \sin^3 \theta) d\theta dt$$

$$= \int_0^{2\pi} \int_0^{\pi/2} -2(\cos t \sin^2 \theta \cos \theta + \sin t \sin^2 \theta \cos \theta + \cos t \sin t \sin^3 \theta) d\theta dt$$

Let's look at the third term in the integrand:

$$\int_0^{2\pi} \int_0^{\pi/2} -2 \cos t \sin t \sin^3 \theta d\theta dt$$

Recognizing that $2 \cos t \sin t = \sin 2t$, we can rewrite this as

$$\int_0^{2\pi} \int_0^{\pi/2} -\sin 2t \sin^3 \theta d\theta dt$$

Evaluating the integral with respect to $t$ first, we get

$$\int_0^{2\pi} -\sin 2t  dt = \left[ \frac{\cos 2t}{2} \right]_0^{2 \pi} = 1 - 1 = 0$$

Similarly, the first and second terms in the integrand involve integrals of $\cos t$ and $\sin t$, respectively, with respect to $t$.  These integrals are also zero.  Thus, the integral of the curl of $\mathbf{F}$ over $\Sigma$ is zero.

## Evaluating the integral using Stokes' theorem

By Stokes' theorem, this is equal to the line integral of $\mathbf{F}$ over the boundary of $\Sigma$:

$$\int_{\Sigma} \nabla \times \mathbf{F} \cdot d\mathbf{S} = \int_{\partial \Sigma} \mathbf{F} \cdot d\mathbf{r}$$

where $d\mathbf{r}$ is the tangent vector to the boundary of $\Sigma$.

The boundary of $\Sigma$ is the circle $x^2 + y^2 = 1, z = 0$. We can parameterize this as

$$\mathbf{r}(t) = \cos t \mathbf{i} + \sin t \mathbf{j}$$

where $0 \leq t \leq 2\pi$.

The line integral of $\mathbf{F}$ over the boundary of $\Sigma$ is

$$\int_{\partial \Sigma} \mathbf{F} \cdot d\mathbf{r} = \int_0^{2\pi} \mathbf{F}(\mathbf{r}(t)) \cdot \mathbf{r}'(t) dt$$

$$= \int_0^{2\pi} \mathbf{F}(\cos t \mathbf{i} + \sin t \mathbf{j}) \cdot (-\sin t \mathbf{i} + \cos t \mathbf{j}) dt$$

$$= \int_0^{2\pi} ((\cos^2 t + \sin^2 t) \mathbf{i} + (\sin^2 t) \mathbf{j} + (\cos^2 t ) \mathbf{k}) \cdot (-\sin t \mathbf{i} + \cos t \mathbf{j}) dt$$

$$= \int_0^{2\pi} (\mathbf{i} + \sin^2 t \mathbf{j} + \cos^2 t \mathbf{k}) \cdot (-\sin t \mathbf{i} + \cos t \mathbf{j}) dt$$

$$= \int_0^{2\pi} (-\sin t + \sin^2 t \cos t) dt$$

$$= \left[ cos t \right]_0^{2\pi} + \left[ \frac{u^3}{3} \right]_0^0$$

$$= 0$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the vector field
def F(x, y, z):
    return np.array([x**2 + y**2, y**2 + z**2, z**2 + x**2])

# Define the surface
def r(t, theta):
    return np.array([np.cos(t)*np.sin(theta), np.sin(t)*np.sin(theta), np.cos(theta)])

# Define the normal vector to the surface
def n(t, theta):
    return np.array([np.cos(t)*np.sin(theta), np.sin(t)*np.sin(theta), np.cos(theta)])

# Define the tangent vector to the boundary of the surface
def dr(t):
    return np.array([-np.sin(t), np.cos(t), np.zeros(t.shape)])

# Define the curl of the vector field
def curlF(x, y, z):
    return np.array([-2*x*z, 2*y*z, 2*x*y])
```

```python
#%matplotlib notebook
npts = 20
```

```python
# Plot the surface
t = np.linspace(0, 2*np.pi, npts)
theta = np.linspace(0, np.pi/2, npts)
T, Theta = np.meshgrid(t, theta)
X, Y, Z = r(T, Theta)
DR = dr(T)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, alpha=0.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
# Plot the normal vector field
N = n(T, Theta)
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, N[0], N[1], N[2], length=0.1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
# Plot the vector field
U, V, W = F(X, Y, Z)
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=0.1)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

```

```python
# Plot the curl of the vector field and the normal vector field
U, V, W = curlF(X, Y, Z)
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=0.1, color='g')
ax.quiver(X, Y, Z, N[0], N[1], N[2], length=0.1, color='r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
# Evaluate the integral of the curl of F over the surface
integral = np.sum(curlF(X, Y, Z)*N)*np.pi/2/npts**2
print(integral)
```

```python
# Plot the vector field and the tangent vector field to the boundary of the surface
t = np.linspace(0, 2*np.pi, npts)
X, Y, Z = r(t, np.ones_like(t)*np.pi/2)
DRX, DRY, DRZ = dr(t)
U, V, W = F(np.cos(t), np.sin(t), np.zeros(t.shape))

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.quiver(X, Y, Z, U, V, W, length=0.2, color='g')
ax.quiver(X, Y, Z, DRX, DRY, DRZ, length=0.2, color='r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set(xlim=(-1, 1), ylim=(-1, 1), zlim=(-1, 1))
plt.show()
```

```python
# Evaluate the integral of F over the boundary of the surface
t = np.linspace(0, 2*np.pi, npts)

integral = np.sum(F(np.cos(t), np.sin(t), np.zeros(t.shape))*dr(t))*2*np.pi/npts

print(integral)
```
