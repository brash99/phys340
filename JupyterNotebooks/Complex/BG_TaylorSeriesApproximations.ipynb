# Taylor Series Approximations

As we discussed in class, the Taylor series is a way to represent a function as an infinite sum of terms. The Taylor series for a function $f(x)$ about a point x=a is given by:

$$f(x) = f(a) + f'(a)(x-a) + \frac{f''(a)(x-a)^2}{2!} + \frac{f'''(a)(x-a)^3}{3!} + ...$$

where $f'(a)$ is the first derivative of $f(x)$ evaluated at $x=a$, $f''(a)$ is the second derivative of $f(x)$ evaluated at $x=a$, and so on. The terms in the Taylor series are given by the $n-$th derivative of $f(x)$ evaluated at $x=a$, divided by $n!$, and multiplied by $(x-a)^n$.

The Taylor series is an infinite series, but we can approximate a function by truncating the series after a finite number of terms. The more terms we include in the series, the better the approximation will be. The Taylor series is particularly useful when we want to approximate a function that is difficult to evaluate directly, but we know the derivatives of the function.

As a first example, let's consider the function $f(x) = sin(x)$. We know that the derivative of $sin(x)$ is $cos(x)$, and the second derivative is $-sin(x)$, and the third derivative is $-cos(x)$, and so on. We can use this information to write down the Taylor series for $sin(x)$ about $x=0$. The Taylor series for $sin(x)$ about $x=0$ is given by:

$$sin(x) = x - x^3/3! + x^5/5! - x^7/7! + ...$$

Written in summation notation, the Taylor series for $sin(x)$ about $x=0$ is:

$$sin(x) = \sum_{n=0}^\infty (-1)^n  \frac{x^{2n+1}}{(2n+1)!}$$

Similary, we can write down the Taylor series for $cos(x)$ about $x=0$:

$$cos(x) = 1 - x^2/2! + x^4/4! - x^6/6! + ...$$

Written in summation notation, the Taylor series for $cos(x)$ about $x=0$ is:

$$cos(x) = \sum_{n=0}^\infty (-1)^n  \frac{x^{2n}}{(2n)!}$$

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# define a function to approximate sin(x) using the Taylor series with n terms
def sin(x, n):
    s = 0
    for i in range(n):
        sign = (-1)**i
        num = x**(2*i+1)
        den = math.factorial(2*i+1)
        s += sign * num / den
    return s

x = np.linspace(-2*np.pi, 2*np.pi, 100)
xd = x*180/np.pi

# Exact value of sin(x)
y = np.sin(x)

# Taylor series up to 2 terms
y2 = sin(x, 2)

# Taylor series up to 5 terms
y5 = sin(x, 5)

# Taylor series up to 10 terms
y10 = sin(x, 10)


# plot sin(x) and its Taylor series approximations
plt.plot(xd, y, label='sin(x)')
plt.plot(xd, y2, label='sin(x) approx n=2')
plt.plot(xd, y5, label='sin(x) approx n=5')
plt.plot(xd, y10, label='sin(x) approx n=10')
plt.legend()
plt.xlabel('x (degrees)')
plt.ylabel('y')
plt.title('Taylor Series Approximation of sin(x)')
plt.ylim(-2, 2)
plt.xlim(-360, 360)

# draw axes
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.show()
```

```python
# Let's look at the x region between 0 and 90 degrees

plt.plot(xd, y, label='sin(x)')
plt.plot(xd, y2, label='sin(x) approx n=2')
plt.plot(xd, y5, label='sin(x) approx n=5')
plt.plot(xd, y10, label='sin(x) approx n=10')
plt.legend()
plt.xlabel('x (degrees)')
plt.ylabel('y')
plt.title('Taylor Series Approximation of sin(x)')
plt.ylim(0, 1.1)
plt.xlim(0, 90)

# draw axes
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.show()
```

```python
# Define a function to approximate cos(x) using the Taylor series with n terms
def cos(x, n):
    s = 0
    for i in range(n):
        sign = (-1)**i
        num = x**(2*i)
        den = math.factorial(2*i)
        s += sign * num / den
    return s

x = np.linspace(-2*np.pi, 2*np.pi, 100)

# Let's test the identity: sin(x)^2 + cos(x)^2 = 1

# Taylor series approximations of sin(x)^2 + cos(x)^2
unity5 = sin(x, 5)**2 + cos(x, 5)**2
unity10 = sin(x, 10)**2 + cos(x, 10)**2

plt.plot(xd, unity5, label='sin(x)^2 + cos(x)^2 approx n=5')
plt.plot(xd, unity10, label='sin(x)^2 + cos(x)^2 approx n=10')

plt.legend()
plt.xlabel('x (degrees)')
plt.ylabel('y')
plt.title('Taylor Series Approximation of sin(x)^2 + cos(x)^2')
plt.ylim(0.5, 1.5)
plt.xlim(-360, 360)
```

```python
# compare small angle approximations

x = np.linspace(0.0, 10.0*np.pi/180.0, 100)
xd = x*180/np.pi

y = np.sin(x)
y1 = sin(x, 1)
y2 = sin(x, 2)
y3 = sin(x, 3)

#y1 = x
#y2 = x - x**3/6.0
#y3 = x - x**3/6.0 + x**5/120.0

# Calculate the error in the small angle approximations
ey1 = np.abs(y - y1)
ey2 = np.abs(y - y2)
ey3 = np.abs(y - y3)

plt.plot(xd, y, label='sin(x)')
plt.plot(xd, y1, label='x')
plt.plot(xd, y2, label='x - x^3/6')
plt.plot(xd, y3, label='x - x^3/6 + x^5/120')

plt.legend()
plt.title('Small Angle Approximations')
plt.xlabel('x (degrees)')
plt.ylabel('y')
```

```python
plt.plot(xd, ey1, label='sin(x) - x')
plt.plot(xd, ey2, label='sin(x) - x + x^3/6')
plt.plot(xd, ey3, label='sin(x) - x + x^3/6 - x^5/120')

plt.legend()
plt.title('Error in Small Angle Approximations')
plt.xlabel('x (degrees)')
plt.ylabel('y')
plt.yscale('log')
```
