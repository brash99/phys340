# The Gamma Function

The gamma function is a generalization of the factorial function to the real numbers. It is defined as:

$$\Gamma(z) = \int_0^\infty t^{z-1}e^{-t}dt$$

for $z > 0$. The gamma function is defined for all complex numbers except for the non-positive integers. The gamma function is a continuous function that interpolates the factorial function. For example, $\Gamma(n) = (n-1)!$ for positive integers $n$.

Sometimes, it is convenient to define the gamma function shifted by one:

$$\Gamma(z+1) = \int_0^\infty t^ze^{-t}dt$$

for $z > -1$. The gamma function satisfies the functional equation:

$$\Gamma(z+1) = z\Gamma(z)$$

for $z > 0$. Also, we then have that $\Gamma(n+1) = (n)!$ for non-negative integers $n$.

The gamma function is a special function that appears in many areas of mathematics, including combinatorics, number theory, and physics. The gamma function is used to define the beta function, which is used in probability theory and statistics.

The gamma function is implemented in the `scipy.special` module as `gamma`. The gamma function is a transcendental function, so it cannot be expressed in terms of elementary functions. 

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.special
import sympy as sp

t = sp.symbols('t')
z = sp.symbols('z', positive=True)

def gamma(z):
    return sp.integrate(t**(z-1)*sp.exp(-t), (t, 0, sp.oo))

gamma(1), gamma(2), gamma(3), gamma(4), gamma(5), gamma(6), gamma(7)
```

```python
fig = plt.figure(figsize=(10, 6))

z_vals = np.linspace(-0.96, 4, 100)
gamma_vals = [sp.gamma(z_val+1) for z_val in z_vals]

plt.plot(z_vals, gamma_vals)
plt.xlabel('z')
plt.ylabel('Gamma(z+1)')
plt.title('Gamma Function')

# Draw axes
plt.axhline(0, color='black', lw=0.5)
plt.axvline(0, color='black', lw=0.5)

# Draw factorial values
for n in range(0, 5):
    plt.scatter(n, sp.gamma(n+1), color='red')
    
plt.grid()

plt.show()
```
