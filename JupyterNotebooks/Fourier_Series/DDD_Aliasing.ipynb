# Aliasing

In Fourier analysis, aliasing is an effect that causes different signals to become indistinguishable (or aliases of one another) when sampled. It also refers to the distortion or artifact that results when the signal reconstructed from samples is different from the original continuous signal.

Aliasing can occur in signals sampled in time, for instance digital audio, and is referred to as temporal aliasing. Aliasing can also occur in spatially sampled signals, for instance moir√© patterns in digital images. Aliasing in spatially sampled signals is called spatial aliasing.

## Example

Let's start by creating a signal sample with a sampling rate of 10 kHz and a signal frequency of 400 Hz. We will then use a complex Fourier series to approximate the signal and plot the Fourier coefficients and the power spectrum.

Main points to note:

- The Nyquist frequency is 5 kHz
- The signal frequency is 400 Hz
- The signal frequency is less than the Nyquist frequency
- The power spectrum should show a peak at the signal frequency
- The Fourier coefficients should be non-zero at the signal frequency
- The Fourier series should be able to approximate the signal

We will consider two cases:

1. We will calculate the Fourier series using a number of terms that corresponds to the Nyquist frequency
2. We will calculate the Fourier series using a number of terms that corresponds to 2.0 times the Nyquist frequency (which is "illegal", but we will do it anyway).

```python
# Aliasing example

import numpy as np
import matplotlib.pyplot as plt

# Create a signal with a sampling rate of 10 kHz
# and a signal frequency of 400 Hz

sampling_rate = 10000 # Hz
signal_freq = 0.04*sampling_rate # = 400 Hz

npts = 100 # number of points - this will give us about four periods of the signal

dt = 1/sampling_rate # s

t = np.linspace(0, npts*dt, npts)

print("Max t = ", t[-1])

f_nyquist = sampling_rate/2 # Hz
print("Nyquist Frequency = ",f_nyquist)

# Create a signal
signal = np.sin(2*np.pi*signal_freq*t)

# Plot the signal
plt.plot(t, signal, '.')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
# Let's use a complex Fourier series to approximate the signal

period = t[-1]
print("Period = ", period)

case = 2

if case == 1:
    nterms = int(1.0*period*f_nyquist)
else:
    nterms = int(2.0*period*f_nyquist)

print("Number of terms = ", nterms)


# Calculate the Fourier coefficients
def cn(n):
    global signal, t
    c = signal * np.exp(-1j * 2 * np.pi * n * t / period)
    return c.sum() / len(t)

def f(t, Nh):
    global signal
    f = np.array([2 * cn(i) * np.exp(1j * 2 * np.pi * i * t / period) for i in range(1, Nh + 1)])
    return f.sum()

yf = np.array([f(tt, nterms) for tt in t])
```

```python
plt.plot(t, yf.real, 'r')
plt.plot(t, signal, 'g.')
plt.xlim([0, 0.01])
```

```python
# Plot the Fourier coefficients

n = np.arange(1, nterms+1)
freq = n/period
c = np.array([cn(i) for i in n])

fig, ax = plt.subplots(1,2, figsize=(10, 5))

ax[0].plot(freq, abs(c.real), 'b')
ax[1].plot(freq, abs(c.imag), 'r')
ax[0].set_xlabel('frequency')
ax[0].set_ylabel('Real(c_n)')
ax[1].set_xlabel('frequency')
ax[1].set_ylabel('Imag(c_n)')
ax[0].set_ylim(0, 0.6)
ax[1].set_ylim(0, 0.6)
ax[0].grid()
```

```python
# Plot the power spectrum

fig, ax = plt.subplots(1,1, figsize=(10, 5))

ax.plot(freq, abs(c)**2, 'b', label='Power')
ax.set_xlabel('frequency')
ax.set_ylabel('Power')
ax.set_ylim(0, 0.6)
ax.grid()

# plot vertical lines at the signal frequency, the Nyquist frequency, and the aliased frequency

ax.axvline(signal_freq, color='r', linestyle='--', label='Signal Frequency')
ax.axvline(f_nyquist, color='g', linestyle='--', label='Nyquist Frequency')
ax.axvline(1.9*f_nyquist, color='k', linestyle='--', label='Aliased Frequency')

ax.legend()
```

```python
f_primary = 0.04*sampling_rate
print(f_primary)
f_alias = 2.06*f_nyquist
print(f_alias)

t_theory = np.linspace(0, period, 100000)

# plot the original signal
fig = plt.figure(figsize=(12, 8))
plt.plot(t, signal, 'bo')

# plot the primary signal
plt.plot(t_theory, np.sin(2.0*np.pi*f_primary*t_theory), 'r')

# plot the alias signal
plt.plot(t_theory, np.sin(2.0*np.pi*f_alias*t_theory), 'g')

plt.xlim([0, 0.0025])
```
