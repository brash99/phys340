# Coupled First Order ODEs

Consider the following system of coupled first order ODEs:

$$\begin{align}
\frac{dx}{dt} =  3x + 4y \\
\frac{dy}{dt} =  4x + 3y
\end{align}$$

where $x$ and $y$ are functions of time $t$.  This system of ODEs can be written in matrix form as:

$$\begin{equation}
\begin{bmatrix}
\frac{dx}{dt}\\
\frac{dy}{dt}
\end{bmatrix}
=
\begin{bmatrix}
3 & 4\\
4 & 3
\end{bmatrix}
\begin{bmatrix}
x\\
y
\end{bmatrix}
\end{equation}$$

From the fundamental theorem for eigenvalues and eigenvectors, we know that we can write:

$$\begin{bmatrix} \frac{dx}{dt}\\ \frac{dy}{dt} \end{bmatrix} = S \Lambda S^{-1} \begin{bmatrix} x\\ y \end{bmatrix}$$


$$S^{-1} \begin{bmatrix} \frac{dx}{dt}\\ \frac{dy}{dt} \end{bmatrix} = \Lambda S^{-1} \begin{bmatrix} x\\ y \end{bmatrix}$$

We can define a new set of variables, $r_1$ and $r_2$, as:

$$\begin{equation}
\begin{bmatrix} r_1\\ r_2 \end{bmatrix} = S^{-1} \begin{bmatrix} x\\ y \end{bmatrix}
\end{equation}$$

Then, the system of ODEs can be written as

$$\begin{equation}
\begin{bmatrix} \frac{dr_1}{dt}\\ \frac{dr_2}{dt} \end{bmatrix} = \Lambda \begin{bmatrix} r_1\\ r_2 \end{bmatrix}
\end{equation}$$

where $\Lambda$ is a diagonal matrix with the eigenvalues of the coefficient matrix on the diagonal.

 

```python
import numpy as np

# Initial conditions

y10 = 1
y20 = 0

# create a coefficient matrix
A = np.array([[3, 4],[4, 3]])
print(A)

# print the eigenvalues and eigenvectors of the coefficient matrix
eigenvalues, eigenvectorsT = np.linalg.eig(A)

# the eigenvectors are the columns of the eigenvectorsT matrix
# taking the transpose of this matrix gives the eigenvectors as rows
eigenvectors = eigenvectorsT.T
print('The eigenvalues are ')
print(eigenvalues)
print('The eigenvectors (columns) are ')
print(eigenvectorsT)
print()
```

## Solution of the System of ODEs

Solving for the eigenvalues and eigenvectors of the coefficient matrix, we find that the eigenvalues are $\lambda_1 = 7$ and $\lambda_2 = -1$.

We can show this by solving the characteristic equation of the coefficient matrix:

$$\begin{equation}
\begin{vmatrix}
3-\lambda & 4\\
4 & 3-\lambda
\end{vmatrix}
= (3-\lambda)(3-\lambda) - 16 = 0
\end{equation}$$

This leads to the characteristic equation $\lambda^2 - 6\lambda - 7 = 0$, which has solutions $\lambda_1 = 7$ and $\lambda_2 = -1$.

The corresponding eigenvectors are $\mathbf{v}_1 = \begin{bmatrix}\frac{1}{\sqrt{2}}\\\frac{1}{\sqrt{2}}\end{bmatrix}$ and $\mathbf{v}_2 = \begin{bmatrix}-\frac{1}{\sqrt{2}}\\\frac{1}{\sqrt{2}}\end{bmatrix}$.

We can show this by solving the equation $(A - \lambda I)\mathbf{v} = 0$ for each eigenvalue.  For $\lambda_1 = 3$, we have

$$\begin{equation}
\begin{bmatrix}
3-7 & 4\\
4 & 3-7
\end{bmatrix}
\begin{bmatrix}
v_{11}\\
v_{21}
\end{bmatrix}
= \begin{bmatrix}
-4 & 4\\
4 & -4
\end{bmatrix}
\begin{bmatrix}
v_{11}\\
v_{21}
\end{bmatrix}
= \begin{bmatrix}
0\\
0
\end{bmatrix}
\end{equation}$$

This leads to the equations $-4v_{11} + 4v_{21} = 0$ and $4v_{11} - 4v_{21} = 0$.  Each of these leads to the condition $v_{21} = v_{11}$.

We can also impose that the eigenvector is normalized, so that $v_{11}^2 + v_{21}^2 = 1$.  This leads to the condition $v_{11}^2 + v_{11}^2 = 1$, or $2v_{11}^2 = 1$, or $v_{11} = \frac{1}{\sqrt{2}}$.  Then, $v_{21} = v_{11} = \frac{1}{\sqrt{2}}$.

For $\lambda_2 = -1$, we have

$$\begin{equation}
\begin{bmatrix}
3-(-1) & 4\\
4 & 3-(-1)
\end{bmatrix}
\begin{bmatrix}
v_{12}\\
v_{22}
\end{bmatrix}
= \begin{bmatrix}
4 & 4\\
4 & 4
\end{bmatrix}
\begin{bmatrix}
v_{12}\\
v_{22}
\end{bmatrix}
= \begin{bmatrix}
0\\
0
\end{bmatrix}
\end{equation}$$

This leads to the equations $4v_{12} + 4v_{22} = 0$ and $4v_{12} + 4v_{22} = 0$.  Each of these leads to the condition $v_{22} = -v_{12}$.

We can also impose that the eigenvector is normalized, so that $v_{12}^2 + v_{22}^2 = 1$.  This leads to the condition $v_{12}^2 + v_{12}^2 = 1$, or $2v_{12}^2 = 1$, or $v_{12} = \frac{1}{\sqrt{2}}$.  Then, $v_{22} = -v_{12} = -\frac{1}{\sqrt{2}}$.

The eigenvectors are then $\mathbf{v}_1 = \begin{bmatrix}\frac{1}{\sqrt{2}}\\\frac{1}{\sqrt{2}}\end{bmatrix}$ and $\mathbf{v}_2 = \begin{bmatrix}\frac{1}{\sqrt{2}}\\-\frac{1}{\sqrt{2}}\end{bmatrix}$.

From these, we create a similar matrix $S$ with the eigenvectors as columns, and a diagonal matrix $L$ with the eigenvalues on the diagonal:

$$\begin{equation}
S = \begin{bmatrix}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}\\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{bmatrix}
\end{equation}$$

$$\begin{equation}
L = \begin{bmatrix}
7& 0\\
0 & -1
\end{bmatrix}
\end{equation}$$

The determinant of the matrix $S$ is $\frac{1}{\sqrt{2}}(-\frac{1}{\sqrt{2}}) - \frac{1}{\sqrt{2}}(\frac{1}{\sqrt{2}}) = -1$.

The inverse of the matrix $S$ is then

$$\begin{equation}
S^{-1} = \frac{1}{-1}\begin{bmatrix}
-\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}\\
-\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}
\end{bmatrix}
\end{equation}$$

or, simplifying, 

$$\begin{equation}
S^{-1} = \begin{bmatrix}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}\\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{bmatrix}
\end{equation}$$

The solution of the system of ODEs is then

$$\begin{equation}
\begin{bmatrix}
x\\
y
\end{bmatrix}
=
S
\begin{bmatrix}
r_1\\
r_2
\end{bmatrix}
=
\begin{bmatrix}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}\\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{bmatrix}
\begin{bmatrix}
r_1\\
r_2
\end{bmatrix}
\end{equation}$$

where $r_1$ and $r_2$ are the solutions of the system of ODEs

$$\begin{equation}
\begin{bmatrix}
\frac{dr_1}{dt}\\
\frac{dr_2}{dt}
\end{bmatrix}
=
\begin{bmatrix}
7& 0\\
0 & -1
\end{bmatrix}
\begin{bmatrix}
r_1\\
r_2
\end{bmatrix}
\end{equation}$$

The solution of this system of ODEs is
    
$$\begin{equation}
\begin{bmatrix}
r_1\\
r_2
\end{bmatrix}
=
\begin{bmatrix}
r_{10} e^{7t}\\
r_{20} e^{-t}
\end{bmatrix}
\end{equation}$$

where $r_{10}$, $r_{20}$  are constants of integration.

We can determine these constants of integration from the initial conditions.  If the initial conditions are $x(0) = 1$, $y(0) = 0$ then the solution of the system of ODEs is

$$\begin{equation}
\begin{bmatrix}
r_{10}\\
r_{20}
\end{bmatrix}
=
\begin{bmatrix}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}\\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{bmatrix}
\begin{bmatrix}
1\\
0
\end{bmatrix}
\end{equation}$$


This leads to the following solution for $y_1$ and $y_2$:

$$\begin{align}
r_{10}= \frac{1}{\sqrt{2}} \\
r_{20}= \frac{1}{\sqrt{2}}
\end{align}$$

The solution of the system of ODEs is then

$$\begin{equation}
\begin{bmatrix}
x\\
y
\end{bmatrix}
=
S
\begin{bmatrix}
r\\
s
\end{bmatrix}
=
\begin{bmatrix}
\frac{1}{\sqrt{2}} & \frac{1}{\sqrt{2}}\\
\frac{1}{\sqrt{2}} & -\frac{1}{\sqrt{2}}
\end{bmatrix}
\begin{bmatrix}
\frac{1}{\sqrt{2}} e^{7t}\\
\frac{1}{\sqrt{2}} e^{-t}
\end{bmatrix}
\end{equation}$$

Multiplying the matrices, we find

$$\begin{equation}
\begin{bmatrix}
x\\
y
\end{bmatrix}
=
\begin{bmatrix}
\frac{1}{\sqrt{2}}(\frac{1}{\sqrt{2}} e^{7t} + \frac{1}{\sqrt{2}} e^{-t})\\
\frac{2}{\sqrt{2}}(\frac{1}{\sqrt{2}} e^{7t} - \frac{1}{\sqrt{2}} e^{-t})
\end{bmatrix}
\end{equation}$$

$$\begin{equation}
\begin{bmatrix}
x\\
y
\end{bmatrix}
=
\begin{bmatrix}
\frac{1}{2} e^{7t} + \frac{1}{2} e^{-t})\\
\frac{1}{2} e^{7t} - \frac{1}{2} e^{-t})
\end{bmatrix}
\end{equation}$$


```python
import sympy as sym

# create a solution matrix
R = sym.Matrix([sym.symbols('r1'), sym.symbols('r2')])
sym.pretty_print(R)

# create a matrix of eigenvectors
S = sym.Matrix(eigenvectorsT)
sym.pretty_print(S)

# create a diagonal matrix of eigenvalues
L = sym.Matrix([[eigenvalues[0], 0], [0, eigenvalues[1]]])

# create the inverse of the matrix of eigenvectors
Sinv = S.inv()
sym.pretty_print(Sinv)

r1 = sym.symbols('r10')*sym.exp(eigenvalues[0] * sym.symbols('t'))
r2 = sym.symbols('r20')*sym.exp(eigenvalues[1] * sym.symbols('t'))

sym.pretty_print(r1)
sym.pretty_print(r2)

Y = S * R

print('The solution is')
sym.pretty_print(Y)

print('r1(0) = ',r1.subs(sym.symbols('t'),0).subs(sym.symbols('t'),0))
print('r2(0) = ',r2.subs(sym.symbols('t'),0).subs(sym.symbols('t'),0))


```

```python
Y0 = S * R.subs(sym.symbols('r1'),r1).subs(sym.symbols('r2'),r2)
Y0 = Y0.subs(sym.symbols('t'),0)
sym.pretty_print(Y0)

Y0dot = S * R.subs(sym.symbols('r1'),sym.diff(r1,sym.symbols('t'))).subs(sym.symbols('r2'),sym.diff(r2,sym.symbols('t')))
Y0dot = Y0dot.subs(sym.symbols('t'),0)
sym.pretty_print(Y0dot)

Y0val = sym.Matrix([sym.symbols('y10'), sym.symbols('y20')])
sym.pretty_print(Y0val)

Y0val = Y0val.subs(sym.symbols('y10'),y10).subs(sym.symbols('y20'),y20)

sym.pretty_print(Y0val)
```

```python
constant_dict = sym.solve([Y0val[0] - Y0[0], Y0val[1] - Y0[1]], [sym.symbols('r10'), sym.symbols('r20')])
```

```python
print(constant_dict[sym.symbols('r10')])
print(constant_dict[sym.symbols('r20')])
```

```python
Y_solution = Y.subs(sym.symbols('r1'),r1).subs(sym.symbols('r2'),r2)
sym.pretty_print(Y_solution)
Y_solution = Y_solution.subs(constant_dict)
sym.pretty_print(Y_solution)
```

```python
import matplotlib.pyplot as plt

t = np.linspace(0, 2, 100)

xt = sym.lambdify(sym.symbols('t'), Y_solution[0], 'numpy')
yt = sym.lambdify(sym.symbols('t'), Y_solution[1], 'numpy')

plt.plot(t, xt(t), label='x')
plt.plot(t, yt(t), label='y')
plt.legend()
plt.show()
```

```python
# now, let's calculate the motion numerically, using the Euler method
# time step
dt = 0.001

# number of time steps
n = 2000

# create arrays to store the solution
tn = np.zeros(n)
x = np.zeros(n)
y = np.zeros(n)
xdot = np.zeros(n)
ydot = np.zeros(n)

# initial conditions
tn[0] = 0
x[0] = 1/np.sqrt(5)
y[0] = -3/np.sqrt(5)

# loop over time steps
for i in range(n-1):
    # calculate the derivatives
    xdot[i] = x[i] + y[i]
    ydot[i] = 4*x[i] + y[i]
    
    # calculate the new values of y1 and y2 (Euler method)
    x[i+1] = x[i] + xdot[i] * dt
    y[i+1] = y[i] + ydot[i] * dt
    
    # calculate the new time
    tn[i+1] = tn[i] + dt
```

```python
# plot the results
fig, ax = plt.subplots(2,figsize=(10, 10))
ax[0].plot(tn, x, 'r.', label='x')
ax[1].plot(tn, y, 'r.', label='y')
ax[0].plot(t, xt(t), 'b-', label='x analytical', linewidth=4)
ax[1].plot(t, yt(t), 'b-', label='y analytical', linewidth=4)
ax[0].set_xlabel('t (s)')
ax[1].set_xlabel('t (s)')
ax[0].set_ylabel('x (m)')
ax[1].set_ylabel('y (m)')
ax[0].legend()
ax[1].legend()
plt.show()
```
