# Mass, Center of Mass, and Moment of Inertia of a Rod of Length L

## Density Function

$$\lambda(x) = \lambda_0 $$

## Calculation of Mass

$$ M = \int \lambda(x) dx $$

$$ = \int \lambda_0  dx $$

$$ = \lambda_0 \int dx $$


$$ = \lambda_0 \left[ x \right]_{-\frac{L}{2}}^{\frac{L}{2}} $$

$$ = \lambda_0 ( \frac{L}{2} - (\frac{L}{2}) = \lambda_0 L $$

$$ \lambda(x) = \frac{M}{L}$$

 Calculation of Moment of Inertia

$$ I_{y} = \int x^2 \lambda(x) dx $$

$$ = \frac{M}{L}\int x^2 dx $$

$$ =  \frac{M}{L} \int x^2 \left( 1 + \frac{x^2}{a^2} \right) dx $$

$$ = \frac{M}{L} \left[ \frac{x^3}{3} \right]_{-\frac{L}{2}}^{\frac{L}{2}} $$

$$ = \frac{M}{L} \left[ \frac{L^3}{24} - (-\frac{L^3}{24}) \right] $$

$$ = \frac{M}{L} \left[ \frac{L^3}{12} \right] $$


$$ = \frac{1}{12} M L^2$$
