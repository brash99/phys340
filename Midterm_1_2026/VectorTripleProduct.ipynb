### Verify the relationship that A dot (B cross C) = B dot (C cross A)
### for the vectors A = (2, -1, 3), B = (-1, 4, 2), and C = (3, 1, -1)
### using sympy





```python

from sympy import symbols, Matrix

# Define the vectors A, B, and C
A = Matrix([2, -1, 3])
B = Matrix([-1, 4, 2])
C = Matrix([3, 1, -1])

# Calculate A dot (B cross C)
B_cross_C = B.cross(C)
A_dot_B_cross_C = A.dot(B_cross_C)
# Calculate B dot (C cross A)
C_cross_A = C.cross(A)
B_dot_C_cross_A = B.dot(C_cross_A)
# Verify the relationship
relationship_verified = A_dot_B_cross_C == B_dot_C_cross_A
print("A dot (B cross C) =", A_dot_B_cross_C)
print("B dot (C cross A) =", B_dot_C_cross_A)
print("Relationship A dot (B cross C) = B dot (C cross A) is", relationship_verified)


```

```python
# Now, prove that this is true for any vectors A, B, and C using sympy

# Define symbolic variables for the components of the vectors
a1, a2, a3 = symbols('a1 a2 a3')
b1, b2, b3 = symbols('b1 b2 b3')
c1, c2, c3 = symbols('c1 c2 c3')

# Define the symbolic vectors A, B, and C
A = Matrix([a1, a2, a3])
B = Matrix([b1, b2, b3])
C = Matrix([c1, c2, c3])

# Calculate A dot (B cross C)
B_cross_C = B.cross(C)
A_dot_B_cross_C = A.dot(B_cross_C)
# Calculate B dot (C cross A)
C_cross_A = C.cross(A)
B_dot_C_cross_A = B.dot(C_cross_A)
# Verify the relationship
relationship_verified = A_dot_B_cross_C - B_dot_C_cross_A
print("A dot (B cross C) =", A_dot_B_cross_C)
print("B dot (C cross A) =", B_dot_C_cross_A)
print("A dot (B cross C) - B dot (C cross A) =", relationship_verified.simplify())
```
