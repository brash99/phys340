```python
import numpy as np
import matplotlib.pyplot as plt

def round_to_half(x):
    return np.round(2*x)/2

N = 100000
sigma = 0.5

T_T, T_R, T_S = 3.5, 3.0, 3.5

eps_T = np.random.normal(0, sigma, N)
eps_R = np.random.normal(0, sigma, N)
eps_S = np.random.normal(0, sigma, N)

S_T = round_to_half(T_T + eps_T)
S_R = round_to_half(T_R + eps_R)
S_S = round_to_half(T_S + eps_S)

S_overall = 0.50*S_T + 0.30*S_R + 0.20*S_S

plt.hist(S_overall, bins=40)
plt.xlabel("Overall Score")
plt.ylabel("Frequency")
plt.title("Monte Carlo Distribution of Overall Score")
plt.show()

print("Mean:", np.mean(S_overall))
print("Std dev:", np.std(S_overall))
```
