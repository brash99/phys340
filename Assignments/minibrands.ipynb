```
import numpy as np
import matplotlib.pyplot as plt
```

```
nfigs = 66
nsamples = 5

iterations = []

for sim in range(10000):
    collected = []
    counter = 0
    while len(collected) < nfigs:
        for i in range(nsamples):
            # choose a random number between 1 and nfigs
            fig = np.random.randint(0, nfigs)
            # if fig is not in collected, add it to the list
            if fig not in collected:
                collected.append(fig)
        # print the list
        #print(collected)
        counter += 1

    # sort the list
    collected.sort()

    # print the list
    #print(collected)

    # print the counter
    #print(counter+1)
    iterations.append(counter)

```

```
# print the expectation value of iterations
print(np.array(iterations).mean())

# calculate a Poisson distribution with this average value
import math
lambda_poisson = np.array(iterations).mean()
x = np.arange(0, 2 * nfigs)
poisson = np.exp(-lambda_poisson) * lambda_poisson ** x / np.array([math.factorial(i) for i in x])
#plt.plot(x, len(iterations) * poisson)
plt.hist(iterations, bins=2 * nfigs, alpha=0.5)
plt.xlim(0, 2 * nfigs)
plt.show()
```
