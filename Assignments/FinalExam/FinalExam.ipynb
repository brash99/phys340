IMPORTANT NOTE!!! The "cylinder" in this problem is not really a cylinder, but rather a shape defined by a curve in the x-z plane that is extended in the y direction. The curve is defined by the equation $z = 2 - x^2$.  See the cell below for a visualization of the surface. 

```python
# Draw the region

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, np.sqrt(1.9999999), 100)
y = np.linspace(0, 4, 100)

# make 3d plot of the region
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
X, Y = np.meshgrid(x, y)
Z = 2 - X**2
ax.plot_surface(X, Y, Z, alpha=0.5)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
```
