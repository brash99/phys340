```
import numpy as np
import matplotlib.pyplot as plt
```

```
# Define rate and hv values
rate = np.array([13, 15, 21, 62, 162])
hv = np.array([600, 620, 650, 680, 700])

rate2 = np.array([15,250,850])
hv2 = np.array([600, 650, 700])
```

```
# Plot the data

plt.figure(figsize=(8, 6))
plt.plot(hv, rate, 'ro-', label='5 uA')
#plt.plot(hv2, rate2, 'bo-', label='15 uA')
plt.xlabel('HV')
plt.ylabel('Average Hits per Event')
plt.title('CDet HV Scan')
# make y axis logarithmic
#plt.yscale('log')
plt.legend()
```

```
words_per_event = rate*2+50
bytes_per_word = 4
bytes_per_event = words_per_event * bytes_per_word
trigger_rate = 3500

# Calculate the data rate
data_rate = trigger_rate * bytes_per_event / 1e6  # in MB/s
for i in range(len(rate)):
    print(f"Rate: {rate[i]} hits/event, Bytes/event: {bytes_per_event[i]} bytes, Data rate: {data_rate[i]:.2f} MB/s")
```
