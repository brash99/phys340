```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
# 4.1

# y limit 1 is y=0, i.e. the intersection of y=0 and y=x
# y limit 2 is y=1
# x limit 1 is x=y
# x limit 2 is x=1

# Draw the graph
x = np.linspace(0, 1, 100)
y1 = 0*np.ones(len(x))
y2 = x

plt.plot(x, y1, label='y=0')
plt.plot(x, y2, label='y=x')
plt.fill_between(x, y1, y2, where=(y2>y1), color='gray', alpha=0.5)
plt.legend()
plt.show()

x = sp.symbols('x')
y = sp.symbols('y')
integrand = x**3*y**4
integral = sp.integrate(integrand, (x, y, 1), (y, 0, 1))

print(integral)
```

```python
# 4.2

# y limit 1 is y=x^2
# y limit 2 is y=x^3
# x limit 1 is x=0, i.e. the first intersection of y=x^2 and y=x^3
# x limit 2 is x=1, i.e. the second intersection of y=x^2 and y=x^3

# Draw the graph
x = np.linspace(0, 1.1, 100)
y1 = x**2
y2 = x**3
plt.plot(x, y1, label='y=x^2')
plt.plot(x, y2, label='y=x^3')
plt.fill_between(x, y1, y2, where=((y1>y2)), color='gray', alpha=0.5)
plt.legend()
plt.show()

x = sp.symbols('x')
y = sp.symbols('y')
integrand = 3*x + 8*y + 1
integral = sp.integrate(integrand, (y, x**3, x**2), (x, 0, 1))

print(integral)
```
