```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
# 7.1

# Define the variables
x = sym.symbols('x', real=True)
y = sym.symbols('y', real=True)

z = x + sym.I*y

# Define the function
f = 2*z**2 + 4*z - 6*sym.I

# Compute the real and imaginary parts
u = sym.re(f)
v = sym.im(f)

print("u = ")
sym.pprint(u.simplify())
print()
print("v = ")
sym.pprint(v.simplify())
print()

# Compute the partial derivatives
ux = sym.diff(u, x)
uy = sym.diff(u, y)
vx = sym.diff(v, x)
vy = sym.diff(v, y)

print("ux = ")
sym.pprint(ux.simplify())
print()
print("uy = ")
sym.pprint(uy.simplify())
print()
print("vx = ")
sym.pprint(vx.simplify())
print()
print("vy = ")
sym.pprint(vy.simplify())
print() 
```

```python
# 7.4

# Define the variables
x = sym.symbols('x', real=True)
y = sym.symbols('y', real=True)
C = sym.symbols('C', real=True)

t = sym.symbols('t', real=True)
s = sym.symbols('s', real=True)
y0 = sym.symbols('y0', real=True)
x0 = sym.symbols('x0', real=True)

# Define the function
u = x**2 - y**2

# Compute the partial derivatives
ux = sym.diff(u, x)
uy = sym.diff(u, y)
uxx = sym.diff(ux, x)
uyy = sym.diff(uy, y)

print("ux = ")
sym.pprint(ux.simplify())
print()
print("uy = ")
sym.pprint(uy.simplify())
print()

print("uxx = ")
sym.pprint(uxx.simplify())
print()
print("uyy = ")
sym.pprint(uyy.simplify())
print()

# Check to see if the function is harmonic
print("uxx + uyy = ")
sym.pprint(uxx + uyy)
print()

# Find the conjugate harmonic function

vy = ux
vx = -uy

first_integrand = ux.subs(y, t)
first_integral = sym.integrate(first_integrand, t)
first_term = first_integral.subs(t, y) - first_integral.subs(t, y0)
print("first term = ")
sym.pprint(first_term.simplify())

second_integrand = -uy.subs(x, s).subs(y, y0)
second_integral = sym.integrate(second_integrand, s)
second_term = second_integral.subs(s, x) - second_integral.subs(s, x0)
print("second term = ")
sym.pprint(second_term.simplify())
print()

v = first_term + second_term + C
v = v.subs(y0, 0).subs(x0, 0)
print("v = ")
sym.pprint(v.simplify())
print()

f = u + sym.I*v
print("f = ")
sym.pprint(f.simplify())
```
